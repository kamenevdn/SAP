## Подпрограммы - бар-коды

```python
def save_png(bar_codes, code_type='ean13'):
    BAR = barcode.get_barcode_class(code_type)
    file_names = []
    for bar_code in bar_codes:
        bar = BAR(bar_code)
        bar.writer=ImageWriter()
        bar.writer.text_distance=1
        filename = bar_code+'.png'
        bar.write(filename,options={"text_distance":1})
        file_names.append(filename)
    return file_names

def save_total_png(pngs, out_file):
    images = list(map(Image.open, pngs))
    width = images[0].size[0]
    height = images[0].size[1]
    total_width = (width+5)*3
    total_height = (len(images)+3)//3*(height+5)
    barcodes_image = Image.new('RGB',(total_width, total_height), color = (255,255,255))
    
    y_offset = 0
    x_offset = 0
    for index, im in enumerate(images):
        barcodes_image.paste(im, (x_offset,y_offset))
        x_offset += width+5
        if (index+1)%3==0:
            x_offset=0
            y_offset += height+5
    barcodes_image.save(out_file)
    for png in pngs:
        os.remove(png)
    return

def meo_barcodes(meo_codes_file, meo_out_file):
    bar_codes = open(meo_codes_file).read().split()
    pngs = save_png(bar_codes, code_type='code128')
    save_total_png(pngs, meo_out_file)
    print('ZIMU: баркоды МЕО записаны в файл -', meo_out_file)
```

## Подпрограммы - создание и комплектация

```python
def read_barcodes(file_shtr):
    """Чтение файла с EAN кодами товаров"""
    with open(out_dir+file_shtr) as f:
        lines = f.readlines()
        barcodes = {}
        for line in lines:
            pos = line.strip().split(',')
            pos = [x.replace('"','') for x in pos]
            barcodes[pos[1]] = pos[0]
    return barcodes



def pack_order(file, meo_code, barcodes, meo_num, partly):
    """Упаковка одного заказа в МЕО"""
    with open(out_dir+file) as f:
        lines = f.readlines()
        order = {}
        order['id'] = ''
        order['items'] = {}
        for index, line in enumerate(lines):
            pos = line.strip().split('","')
            pos = [x.replace('"','') for x in pos]
            if order['id'] == '':
                order['id'] = pos[0]
            order['items'][str(index)] = {}
            order['items'][str(index)]['item_id'] = pos[1]
            order['items'][str(index)]['item_art'] = pos[2]
            order['items'][str(index)]['item_text'] = pos[3]
            order['items'][str(index)]['item_qnt'] = pos[4]
            
    #Формируем список допустимых кодов МЕО по количеству из meo_num
    #из которого потом случайным образом будем доставать код МЕО
    meo_codes = []
    for num in range(meo_num):
        meo_codes.append('MEO'+str(int(meo_code.split('MEO')[1])+num))
        
    out_file_name = in_dir+'Уп_зак_GOOD_[УПАКОВКА_ЗАКАЗОВ]SSSSS'+order['id']+'.DAT'
    with open(out_file_name, 'w') as f:
        current_time = datetime.now()
        for index, pos in enumerate(order['items']):
            for one_item in range(int(order['items'][str(index)]['item_qnt'])):
                #Если делаем частичную комплектацию (partly), то случайным образом пропускаем товары
                if (partly and np.random.randint(0,5)==0):
                    continue
                    
                write_string = order['id']+','+\
                               meo_codes[np.random.randint(0,meo_num)]+','+\
                               barcodes[order['items'][str(index)]['item_id']]+','+\
                               order['items'][str(index)]['item_id']+','+\
                               '1'+','+\
                               current_time.strftime("%Y/%m/%d %H:%M:%S")+'\n'
                               #current_time.strftime("%d.%m.%y %H:%M:%S")+'\n'
                               
                f.write(write_string)
                current_time = current_time + timedelta(seconds=1)
    return



def pack(files_pack, meo_code, barcodes, meo_num, partly):
    for file in files_pack:        
        pack_order(file, meo_code, barcodes, meo_num, partly)
        #На заказ резервируем по 2 кода МЕО для тестов
        meo_code = 'MEO'+str(int(meo_code.split('MEO')[1])+meo_num)
    new_free_meo_code = meo_code
    return new_free_meo_code



def del_pack_files(files_pack, file_shtr):
    """Удаляем файлы упаковки из папки OUT"""
    for file in files_pack:
        os.remove(out_dir+file)
    os.remove(out_dir+file_shtr)

    
    
def read_meo(meo_file, plant):
    """Читаем номер МЕО из файла"""
    with open('meo_number.txt') as f:
        meo_code = f.read().replace('XXXX', plant)
    return meo_code



def write_meo(meo_file, new_free_meo_code):
    with open(meo_file, 'w') as f:
        new_free_meo_code = 'MEOXXXX'+new_free_meo_code[7:]
        f.write(new_free_meo_code)
    print('ZIMKN: Следующий свободный код MEO (записан в файл): ',new_free_meo_code)
    return new_free_meo_code



def tsd(plant, meo_file, file_shtr, meo_num=2, partly=0):
    """Эмуляция сборки заказа на ТСД.
    Чтение выходных файлов и формирование входных файлов с полной комплектацией
    meo_num - кол-во МЕО для упаковки, номера присваиваются случайно
    partly - 0 или 1. Флаг частичной сборки. Если 1 - то при комплектации случайным образом могут быть
             пропущены товары
    """
    files_pack    = [f for f in os.listdir(out_dir) if re.match(r'\[УПАКОВКА_ЗАКАЗОВ\]SSSSS.*\.DAT', f)]
    
    #Начальный код MEO (после использования в файл запишется следующий свободный номер)
    meo_code = read_meo(meo_file, plant)
    
    #Читаем файл со штрих-кодами
    barcodes = read_barcodes(file_shtr)
    
    #Пакуем заказы в MEO и записываем в файл следующий незанятый код MEO
    new_free_meo_code = pack(files_pack, meo_code, barcodes, meo_num, partly)
    write_meo(meo_file, new_free_meo_code)
    
    #Удаляем исходные файлы упаковкиа
    del_pack_files(files_pack, file_shtr)
    
    
def create_zimg_num(nd_template_path, system_id, order_num):
    orders_created = []
    for num in range(order_num):
        order_created = zimg(nd_template_path, system_id)
        orders_created.append(order_created)
    #Отфильтровываем несозданные заказы и проверяем, создались ли заказы
    orders_created = list(filter(lambda a: a != None, orders_created))
    if len(orders_created)==0:
        raise ValueError('Заказы не созданы, проверьте ошибки.')
    return orders_created
        
    
def zimg(template_path, system='ERD'):
    """Создание zimg заказов из Excel шаблона (строго по одному за раз!)
    в программе Z_IM_CREATE_ORDERS"""
      
    #Подставляем в шаблон код системы ERD/ERT
    template_path = template_path.replace('XXXX',system)
    #Обновляем шаблон новыми номерами заказов
    template = pd.read_excel(template_path,dtype=str)
    template = template.replace('nan','')
    
    #Читаем и записываем следующий номер заказа
    order_num_file_syst = order_num_file.replace('XXXX',system)
    with open(order_num_file_syst) as f:
        order_num = int(f.readline())
        order_num_new = order_num+1
    with open(order_num_file_syst,'w') as f:
        f.write(str(order_num_new))

    #template['ORDER_ID'] = template['ORDER_ID'].apply(int)+len(template['ORDER_ID'].unique())
    template['ORDER_ID'] = order_num
    template.to_excel(template_path,index=False)
    
    try:   
        sap = ss.sap()
        session = sap.session()
        
        sap.run('se38')
        session.findById("wnd[0]/usr/ctxtRS38M-PROGRAMM").text = "Z_IM_CREATE_ORDERS"
        session.findById("wnd[0]").sendVKey(8)
        
        session.findById("wnd[0]/usr/ctxtP_FILE").text = template_path
        session.findById("wnd[0]/usr/radP_CREATE").select()
        session.findById("wnd[0]/usr/radP_CREATE").setFocus()
        session.findById("wnd[0]").sendVKey(8)
        screen = session.findById("wnd[0]/usr")
        for response in list(screen.children)[2:]:
            if response.text != '':
                None
                #print(response.text)
        #print()
        session.findById("wnd[0]").sendVKey(3)
        
        session.findById("wnd[0]/usr/ctxtP_FILE").text = template_path
        session.findById("wnd[0]/usr/radP_COCKP").select()
        session.findById("wnd[0]/usr/radP_COCKP").setFocus()
        session.findById("wnd[0]").sendVKey(8)
        screen = session.findById("wnd[0]/usr")
        for response in list(screen.children)[2:]:
            if response.text != '':
                None
                #print(response.text)
        print(list(screen.children)[-1].text)
        order = response.text.split(' ')[-1]
        
        if order.isdigit():
            return order
    except:
        sap.error_print()
        
    finally:
        session = None

        
        
def upload_to_tsd():
    """Выгрузка заказов в ТСД в реестре заказов в кокпите"""
    try:
        sap = ss.script()
        session = sap.session()
        session.findById("wnd[0]/usr/cntlALV_0100/shellcont/shellcont/shell/shellcont[0]/shell").pressToolbarButton("EXPORT")
        time.sleep(1)
        print('ZIMKN: Заказы выгружены в ТСД')
    except:
        sap.error_print()
        
    finally:
        session = None
    
    
    
def start_cockpit(transaction, orders_created, plant):
    """Запуск тр.кокпита с указанным списком заказов и заводом"""
    try:
        sap = ss.script()
        session = sap.session()
        sap.run(transaction)
        session.findById("wnd[0]/usr/ctxtS_WRK-LOW").text = plant
        
        #Вставляем из буфера заказы
        pyperclip.copy('\r\n'.join(orders_created))
        session.findById("wnd[0]/usr/btn%_S_VB_%_APP_%-VALU_PUSH").press()
        session.findById("wnd[1]/tbar[0]/btn[16]").press()
        session.findById("wnd[1]/tbar[0]/btn[24]").press()
        session.findById("wnd[1]/tbar[0]/btn[8]").press()
        session.findById("wnd[0]").sendVKey(8)
    except:
        sap.error_print()
    finally:
        session = None


        
def zimkn_approve_compl(partly):
    """Загружаем из ТСД и подтверждаем комплектацию"""
    try:
        sap = ss.script()
        session = sap.session()
        session.findById("wnd[0]/usr/cntlALV_0100/shellcont/shellcont/shell/shellcont[0]/shell").pressToolbarButton("IMPORT")
        session.findById("wnd[0]/usr/cntlALV_0100/shellcont/shellcont/shell/shellcont[1]/shell").pressToolbarButton("SAVE_P")
        try:
            session.findById("wnd[1]/usr/btnBUTTON_1").press()
        except:
            None
        if partly:
            print('ZIMKN: Заказы частично скомплектованы и подтверждены.')
        else:
            print('ZIMKN: Заказы скомплектованы и подтверждены.')

    except:
        sap.error_print()
    finally:
        session = None
        
        
def create_ND_hybris_order(delivery_date, plant, materials, mat_quantity=1):
    """Создание Nextday заказов в Hybris"""
    with req.Session() as session:
        headers = {"content-type": "application/json",
               "Referer": "https://go.detmir-stage.ru/cart/",
               "Sec-Fetch-Mode": "cors",      
               "User-Agent":"Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36",
               "X-CSRF-Token": "null"}
        
        self_req = session.get(
            url = 'https://api.go.detmir-stage.ru/v1/users/self',
            headers = headers).json()
        
        cart_id = self_req['carts'][0]['href']
        user_id = self_req['carts'][0]['id']
        
        items_url = 'https://api.go.detmir-stage.ru/v2'+cart_id+'/items'
        
        for material in materials:
            #Кладем товары в корзину
            items = session.post(url = items_url, json = {"id":material}, headers = headers).json()
            
            #Выбираем метод доставки
            courier = session.get(url = 'https://api.go.detmir-stage.ru/v1/'+cart_id+'?filter=delivery.address.code:61000001000;delivery.type:courier&exclude=delivery&warehouseCode='+plant+'&deliveryCourierMethod=NEXTDAY')
            items_plus_url = 'https://api.go.detmir-stage.ru/v1/'+cart_id+'/items/'+material  
            
            #Увеличиваем кол-во в корзине
            items_plus = session.put(url = items_plus_url, json = {"quantity":mat_quantity}, headers = headers).json()
 
        order = session.post(
            url = 'https://api.go.detmir-stage.ru/v1/orders',
            json = {"agreed":True,
                    "notification":{"sms":True,"phone":False},
                    "address":{"street":"ул Дарвина",
                               "house":"27",
                               "city":"Ростов-на-Дону",
                               "code":"61000001000",
                               "floor":"1",
                               "entrance":"1",
                               "apartment":"1",
                               "geoLat":47.241596,
                               "geoLon":39.6327646},
                    "contact":{"customer":"Каменев Дмитрий",
                               "phone":"+79031784239",
                               "email":"dkamenev@prota.ru"},
                    "delivery":{"type":"courier",
                                "store":{"id":None},
                                "date": delivery_date,
                                "period":{"from":delivery_date+"T09:00:00Z",
                                          "to":delivery_date+"T18:00:00Z",
                                          "code":1},
                                "warehouseCode":int(plant),
                                "courierMethod":"NEXTDAY"},
                    "payments":[{"type":"cash"}],
                    "comment":"Тестовый заказ",
                    "bonus_card":None},
            headers = headers).json()
    print(order['id'])
    return order['id']

def create_LM_hybris_order(delivery_date, plant, materials, mat_quantity=1):
    """Создание Nextday заказов в Hybris"""
    with req.Session() as session:
        headers = {"content-type": "application/json",
               "Referer": "https://go.detmir-stage.ru/cart/",
               "Sec-Fetch-Mode": "cors",      
               "User-Agent":"Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36",
               "X-CSRF-Token": "null"}
        
        self_req = session.get(
            url = 'https://api.go.detmir-stage.ru/v1/users/self',
            headers = headers).json()
        
        cart_id = self_req['carts'][0]['href']
        user_id = self_req['carts'][0]['id']
        
        items_url = 'https://api.go.detmir-stage.ru/v2'+cart_id+'/items'
        
        for material in materials:
            #Кладем товары в корзину
            items = session.post(url = items_url, json = {"id":material}, headers = headers).json()
            
            #Выбираем метод доставки
            courier = session.get(url = 'https://api.go.detmir-stage.ru/v1/'+cart_id+'?filter=delivery.address.code:77000000000;delivery.type:courier&exclude=delivery&warehouseCode='+plant+'&deliveryCourierMethod=EXPRESS')
            
            items_plus_url = 'https://api.go.detmir-stage.ru/v1/'+cart_id+'/items/'+material  
            
            #Увеличиваем кол-во в корзине
            items_plus = session.put(url = items_plus_url, json = {"quantity":mat_quantity}, headers = headers).json()
 
        order = session.post(
            url = 'https://api.go.detmir-stage.ru/v1/orders',
            json = {"agreed":True,
                    "notification":{"sms":True,"phone":False},
                    "address":{"street":"ул Арбат",
                               "house":"1",
                               "city":"Москва",
                               "code":"77000000000",
                               "floor":"1",
                               "entrance":"1",
                               "apartment":"1",
                               "geoLat":55.7519074,
                               "geoLon":37.5992175},
                    "contact":{"customer":"Каменев Дмитрий",
                               "phone":"+79031784239",
                               "email":"dkamenev@prota.ru"},
                    "delivery":{"type":"courier",
                                "store":{"id":None},
                                "date": delivery_date,
                                "period":{"from":delivery_date+"T11:00:00Z",
                                          "to":delivery_date+"T14:00:00Z",
                                          "code":4},
                                "warehouseCode":int(plant),
                                "courierMethod":"EXPRESS"},
                    "payments":[{"type":"cash"}],
                    "comment":"Тестовый заказ",
                    "bonus_card":None},
            headers = headers).json()
    print(order['id'])
    return order['id']
```

## Подпрограммы - палетирование, отгрузка

```python
def start_zimu(plant, agent, date):
    sap = ss.sap()
    try:
        session = sap.session()
        sap.run('zimu')
        session.findById("wnd[0]/usr/ctxtS_WERKS-LOW").text = plant
        session.findById("wnd[0]/usr/ctxtS_AGENT-LOW").text = agent
        session.findById("wnd[0]/usr/ctxtS_ADDTI-LOW").text = date
        session.findById("wnd[0]/tbar[1]/btn[8]").press()
    except:
        sap.error_print()
    finally:
        session = None

def zimu_into_tsd():
    sap = ss.sap()
    try:
        session = sap.session()
        session.findById("wnd[0]/usr/cntlALV_0100/shellcont/shell").pressToolbarButton("EXPORT")
        print('ZIMU: Данные выгружены в ТСД.')
    except:
        sap.error_print()
    finally:
        session = None

def tsd_read_pallet_file():
    """Читаем выгруженные файлы паллетирования
    и возвращаем словарь с ключами-идентификаторами вида 'CSE_20190903' для каждого считанного файла"""
    files_pallet    = [f for f in os.listdir(out_dir) if re.match(r'\[Паллетирование_NextDay\].*\.DAT', f)]
    #Читаем выгруженные файлы паллетирования
    agent_info = {}
    for file in files_pallet:
        file_id = file.split('.')[0].split(']')[1]
        with open(out_dir+file) as f:
            meo_lines = f.readlines()
        pallet_info = []
        for meo_line in meo_lines:
            info = {}
            meo_line_split = meo_line.split('","')
            info['id'] = meo_line_split[0].replace('"','')
            info['meo_id'] = meo_line_split[1]
            info['meo_task'] = meo_line_split[2]
            info['order_id'] = meo_line_split[4]
            info['address'] = meo_line_split[12].replace('"','')
            pallet_info.append(info)
        agent_info[file_id] = pallet_info
    return agent_info

def tsd_create_pallet_file(agent_info, pal_num):
    """Формируем файл для загрузки из ТСД"""
    current_time = datetime.now()
    meo_codes = []
    for agent_id in list(agent_info.keys()):
        out_file_name = in_dir+'EO_PALET_[Паллетирование_NEXTDAY]'+agent_id+'.DAT'
        write_text = []
        
        for meo_line in agent_info[agent_id]:
            #Пропускаем все МЕО, которые не надо паллетировать
            if meo_line['meo_task'] != 'Разместить на палету':
                continue
            pallet_num = np.random.randint(1, pal_num+1)  #Условный номер палеты, случайно выбираем из кол-ва pal_num
            
            write_string = '"'+meo_line['meo_id']                          +'"'+','+\
                           '"'+str(pallet_num)                             +'"'+','+\
                           '"'+meo_line['order_id']                        +'"'+','+\
                           '"'+current_time.strftime("%Y/%m/%d %H:%M:%S")  +'"'+','+\
                           '"'+meo_line['meo_task']                        +'"'+','+\
                           '"'+meo_line['address']                         +'"'+'\n'
            write_text.append(write_string)
            current_time = current_time + timedelta(seconds=1)
            meo_codes.append(meo_line['meo_id'])
        
    
    
        #Если была хотя бы одна МЕО для паллетирования, то создаем выходной файл
        if len(write_text)>0:
            with open(out_file_name, 'w') as f:
                for write_line in write_text:
                    f.write(write_line)
            print('ZIMU:', agent_id,'- Cформирован файл для загрузки в SAP:',out_file_name)
        else:
            print('ZIMU:', agent_id,'- Выходной файл из ТСД не сформирован, т.к. нет МЕО для паллетирования!')
    
    #Сохраняем файл со списком выгруженных МЕО и генерим картинку с баркодами этих МЕО
    if len(meo_codes) > 0:
        with open(meo_codes_file, 'w') as file:
            for code in meo_codes:
                file.write("%s\n" % code)
        print('ZIMU: записан файл с выгруженными МЕО - ', meo_codes_file)
               
        #Сохраняем баркоды МЕО в файле
        meo_barcodes(meo_codes_file, meo_out_file)  
    
def tsd_pallet_read_create(pal_num):
    """Читаем выходной файл и формируем файл для загрузки из ТСД"""
    agent_info = tsd_read_pallet_file()
    tsd_create_pallet_file(agent_info, pal_num)
    
    
def zimu_from_tsd_save():
    sap = ss.sap()
    try:
        session = sap.session()
        session.findById("wnd[0]/usr/cntlALV_0100/shellcont/shell").pressToolbarButton("IMPORT")
        session.findById("wnd[0]/usr/cntlALV_0100/shellcont/shell").pressToolbarButton("SAVE")
        print('ZIMU: Данные палетирования сохранены.')
    except:
        sap.error_print()
    finally:
        session = None
        
def zimp_issue(plant, agent):
    sap = ss.sap()
    try:
        session = sap.session()
        sap.run('zimu')
        session.findById("wnd[0]/usr/ctxtS_WERKS-LOW").text = plant
        session.findById("wnd[0]/usr/ctxtS_AGENT-LOW").text = agent
        session.findById("wnd[0]/tbar[1]/btn[8]").press()
        grid = session.findById("wnd[0]/usr/cntlALV_0100/shellcont/shell")
        row_count = grid.RowCount
        selected_Rows = "0-"+str(row_count-1)
        session.findById("wnd[0]/usr/cntlALV_0100/shellcont/shell").currentCellColumn = ""
        session.findById("wnd[0]/usr/cntlALV_0100/shellcont/shell").selectedRows = selected_Rows
        session.findById("wnd[0]/usr/cntlALV_0100/shellcont/shell").pressToolbarButton("SAVE")
        session.findById("wnd[1]/usr/txtZIMK_SHIPPING_SCR_0200-TEXT1").text = "Иванов Водитель Транспортович"
        session.findById("wnd[1]/usr/txtZIMK_SHIPPING_SCR_0200-TEXT2").text = "999333999"
        session.findById("wnd[1]/usr/ctxtG_DATE0200").text = "010119"
        session.findById("wnd[1]/usr/txtZIMK_SHIPPING_SCR_0200-SIGNI").text = "Е001КХ"
        session.findById("wnd[1]/usr/txtZIMK_SHIPPING_SCR_0200-EXTI1").text = "Mercedes"
        session.findById("wnd[1]/tbar[0]/btn[0]").press()
        print('ZIMP: Товары отгружены Агенту.')
        sap.read_alv(grid)
    except:
        sap.error_print()
    finally:
        session = None
```
